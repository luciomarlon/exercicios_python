```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
planilha = pd.read_excel(r"pessoas.xlsx")
```

```python
planilha.head()
```

```python
   plt.hist(planilha['Idade'], bins = 20) #histograma
   plt.show()
```

```python
#platando proporcionalmente por nome
planilha.groupby(['Nome']).Idade.count().sort_values()[-7:].plot(kind = "pie", autopct='%0.2f%%', label = 'Idade Pessoas');
```

```python
#plotando proporcionalmente por idade
planilha.groupby(['Idade']).Idade.count().sort_values()[-7:].plot(kind = "pie",
                                                                        autopct = '%1.2f%%',
                                                                        figsize=(5,5),
                                                                        label = 'Idades por Pessoas');
```
